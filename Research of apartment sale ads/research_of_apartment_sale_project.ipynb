### Посчитайте и добавьте в таблицу новые столбцы

Добавим следующие столбцы в таблицу:
* price_one_square_meter - цена одного квадратного метра;
* exposition_weekday - день недели публикации объявления (0 — понедельник, 1 — вторник и так далее);
* exposition_month - месяц публикации объявления;
* exposition_year - год публикации объявления;
* floor_category - тип этажа квартиры (значения — «‎первый», «последний», «другой»);
* city_centers_nearest_km - расстояние до центра города в километрах

```python
# рассчитаем цену одного квадратного метра и округляем ее
data['price_one_square_meter'] = (data['last_price'] / data['total_area']).round(2)
```

```python
data['exposition_weekday'] = data['first_day_exposition'].dt.weekday # добавим столбец с днем недели публикации объявления

data['exposition_month'] = data['first_day_exposition'].dt.month # добавим столбец с месяцем публикации объявления

data['exposition_year'] = data['first_day_exposition'].dt.year # добавим столбец с годом публикации объявления
```

```python
# Вводим функцию для категоризации этажности в доме
def get_floor_category(row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == 1:
        return 'первый'
    elif floor == floors_total:
        return 'последний'
    else:
        return 'другой'
    
data['floor_category'] = data.apply(get_floor_category, axis=1) # добавляем столбец с категорией этажа квартиры
```

```python
data['city_centers_nearest_km'] = round(data['city_centers_nearest'] / 1000) 
# добавляем столбец с расстоянием до центра города в км
```

* Категория "0" квартиры стоимостью до 3_400_000;
* Категория "3" квартиры стоимостью от 3_400_000 до 4_600_000;
* Категория "4" квартиры стоимостью от 4_600_000 до 6_800_000;
* Категория "6" квартиры стоимостью от 6_800_000;
* Категория "-1" квартиры в которых стоимость не указана;

```python
# вводим функцию для категоризации цены
def price_category(n):
    try:
        if n <= 3400000:
            return '0'
        elif n <= 4600000:
            return '3'
        elif n <= 6800000:
            return '4'
        elif n > 6800000:
            return '6'
    except:
        return '-1'
    
price_category('NaN') # проверяем функцию
```

```python
df['last_price_cat'] = df['last_price'].apply(price_category) # добавляем столбец
df.head(10) # проверяем созданный столбец
```
