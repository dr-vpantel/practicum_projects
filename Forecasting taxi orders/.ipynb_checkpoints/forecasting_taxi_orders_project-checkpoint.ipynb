#  Прогнозирование заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Задача: построить модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

План работы:

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.


Данные лежат в файле `taxi.csv`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

## Обучение

### Подготовка

```python
# cтроим график автокорреляции
fig, ax = plt.subplots(figsize=(10, 5))  
plot_acf(df['num_orders'].dropna(), lags=50, ax=ax)  

ax.set_title("Автокорреляционная функция (ACF)", fontsize=14)
ax.set_xlabel("Лаги", fontsize=12)
ax.set_ylabel("Коэффициент автокорреляции", fontsize=12)
ax.grid(True)

plt.show()
```

```python
# напишем функцию для подготовки признаков
def make_features(data, rolling_mean_size):
    data = pd.DataFrame(data)
    data['hour'] = data.index.hour  
    data['dayofweek'] = data.index.dayofweek  

    # оставляем только важные лаги
    important_lags = [1, 2, 3, 4, 5, 24, 25]
    for lag in important_lags:
        data[f'lag_{lag}'] = data['num_orders'].shift(lag)

    # добавляем скользящее среднее
    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()

    return data

df = make_features(df['num_orders'], rolling_mean_size=30)
df.head()
```

```python
# создание выборок и подготовка данных для обучения
RANDOM_STATE = 42

train, test = train_test_split(df, 
                               shuffle=False, 
                               test_size=0.2, 
                               random_state = RANDOM_STATE)
train = train.dropna()

X_train = train.drop('num_orders', axis=1)
y_train = train['num_orders']
X_sample = test.drop('num_orders', axis=1)
y_sample = test['num_orders']

X_valid, X_test, y_valid, y_test = train_test_split(X_sample,
                                                    y_sample, 
                                                    shuffle=False, 
                                                    test_size=0.5, 
                                                    random_state = RANDOM_STATE)

print('Тренировочная обучающая выборка:', X_train.shape[0], 'Целевой признак:', y_train.shape[0])
print('Валидационная обучающая выборка:', X_valid.shape[0], 'Целевой признак:', y_valid.shape[0])
print('Тестовая обучающая выборка:', X_test.shape[0], 'Целевой признак:', y_test.shape[0])
```

```python
# смотрим крайние даты
print(X_train.index.min(), X_train.index.max())
print(X_valid.index.min(), X_valid.index.max())
print(X_test.index.min(), X_test.index.max())
```

```python
# функция для подбора наилучших гиперпараметров
def train_models_with_random_search(model, params, X_train, y_train, random_state=RANDOM_STATE):
    search = RandomizedSearchCV(
        model,
        param_distributions=params,
        scoring='neg_root_mean_squared_error',
        cv=TimeSeriesSplit(n_splits=3),
        n_iter=5, 
        random_state=random_state,
        n_jobs=-1
    )
    
    search.fit(X_train, y_train)
    
    best_rmse = round(abs(search.best_score_), 2) 
    best_params = search.best_params_
    print(f'Лучшая RMSE модели {model.__class__.__name__}: {best_rmse}')
    print(f'Лучшие гиперпараметры {model.__class__.__name__}: {best_params}')
    
    return best_params
```

```python
# функция для обучения модели с подобранными гиперпараметрами
def trained_model(model, params, X_train, y_train, X_valid, y_valid, random_state=RANDOM_STATE):
    model.set_params(**params)
    model.fit(X_train, y_train)
    predictions = model.predict(X_valid)
    
    mse = mean_squared_error(y_valid, predictions)
    rmse = round(np.sqrt(mse), 2)
    print(f'RMSE модели {model.__class__.__name__} на валидационной выборке : {rmse}')
    
    plt.figure(figsize=(12, 6))
    plt.plot(y_valid.index, y_valid.values, label='Истинные значения')
    plt.plot(y_valid.index, predictions, label='Предсказанные значения')
    
    plt.xlabel('Дата')
    plt.ylabel('Количество заказов')
    plt.title(f'Сравнение предсказаний и истинных значений ({model.__class__.__name__})')
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.show()
    
    return model, rmse
```

### Модель линейной регрессии

```python
params = {'fit_intercept': [True, False]}

params_lr = train_models_with_random_search(LinearRegression(), params, X_train, y_train)
```

```python
model_lr, rmse_lr = trained_model(LinearRegression(), params_lr, X_train, y_train, X_valid, y_valid)
```

### Модель LGBMRegressor 

```python
params = {'n_estimators': [100, 150, 200],  'learning_rate': [0.03, 0.05, 0.07],  'max_depth': [3, 5, 7], 'verbose': [-1]}

params_lgbmr = train_models_with_random_search(LGBMRegressor(), params, X_train, y_train)
```

```python
model_lgbmr, rmse_lgbmr = trained_model(LGBMRegressor(), params_lgbmr, X_train, y_train, X_valid, y_valid)
```

### Модель CatBoostRegressor

```python
params = {'iterations': [100, 200, 300], 'learning_rate': [0.01, 0.03, 0.05], 'depth': [4, 6, 8]}

catboost_model = CatBoostRegressor(silent=True)

params_cbr = train_models_with_random_search(catboost_model, params, X_train, y_train)
```

```python
model_cbr, rmse_cbr = trained_model(catboost_model, params_cbr, X_train, y_train, X_valid, y_valid)
```

### Проверка модели на адекватность

```python
# константная модель
dummy_model = DummyRegressor(strategy="mean")
dummy_model.fit(X_train, y_train)
dummy_predicted = dummy_model.predict(X_valid)
rmse_dummy = mean_squared_error(y_valid, dummy_predicted)**0.5

print('RMSE модели DummyRegressor:', rmse_dummy)
```

```python
# модель прогнозирующая по предыдущему значению ряда
pred_previous = y_test.shift(fill_value=y_train.iloc[-1])
rmse_pp = mean_squared_error(y_test, pred_previous)**0.5
print('RMSE по предыдущему значению ряда:', rmse_pp)
```

### Итоги обучения

```python
models = pd.DataFrame({
    'Модель': ['Линейная регрессия', 'LGBMRegressor ', 'CatBoostRegressor', 'DummyRegressor', 'Предыдущее значение ряда'],
    'RMSE': [rmse_lr, rmse_lgbmr, rmse_cbr, rmse_dummy, rmse_pp]
})

models
```

**Вывод:** метрика RMSE меньше 48 (условия проекта) наблюдается у 3 моделей: модель линейной регрессии (35.89), LGBMRegressor (32.33), CatBoostRegressor (32.37). Метрики моделей меньше значений RMSE DummyRegressor и вычисленной RMSEпо предыдущему значению ряда.
Оценим результат наилучшей модели (LGBMRegressor) на тестовой выборке.

## Тестирование

Посмотрим результат всех трех моделей на тестовой выборке.

```python
# объединим тренировочную и тестовую выборки
X_train = pd.concat([X_train, X_valid], ignore_index=True)
y_train = pd.concat([y_train, y_valid], ignore_index=True)
```

```python
# функция для обучения модели с подобранными гиперпараметрами
def test_model(model, params, X_train, y_train, X_test, y_test, random_state=RANDOM_STATE):
    model.set_params(**params)
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    
    mse = mean_squared_error(y_test, predictions)
    rmse = round(np.sqrt(mse), 2)
    print(f'RMSE модели {model.__class__.__name__} на тестовой выборке : {rmse}')
    
    plt.figure(figsize=(12, 6))
    plt.plot(y_test.index, y_test.values, label='Истинные значения')
    plt.plot(y_test.index, predictions, label='Предсказанные значения')
    
    plt.xlabel('Дата')
    plt.ylabel('Количество заказов')
    plt.title(f'Сравнение предсказаний и истинных значений ({model.__class__.__name__})')
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.show()
    
    return
```

```python
test_model(LGBMRegressor(), params_lgbmr, X_train, y_train, X_test, y_test)
```

**Вывод:** RMSE на тестовой выборке модели LGBMRegressor составила 43.63 (соответсвует условиям проекта).
