# Сборный проект

**Описание проекта**

Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Вам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.
Перед вами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и вы планируете кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируете ли вы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.
В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

**План работы**
1. Изучение общей информации
2. Предобработка данных
3. Исследовательский анализ данных
4. Составлние портрета пользователя каждого региона
5. Общий вывод

**Описание данных**

Name — название игры

Platform — платформа

Year_of_Release — год выпуска

Genre — жанр игры

NA_sales — продажи в Северной Америке (миллионы проданных копий)

EU_sales — продажи в Европе (миллионы проданных копий)

JP_sales — продажи в Японии (миллионы проданных копий)

Other_sales — продажи в других странах (миллионы проданных копий)

Critic_Score — оценка критиков (максимум 100)

User_Score — оценка пользователей (максимум 10)

Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

Данные за 2016 год могут быть неполными.

```python
data.columns = data.columns.str.lower()  # приводим названия столбцов к нижнему регистру
data.head()  # проверяем
```

```python
# проводим проверку на дубликаты по выбранным столбцам
duplicated = data[data.duplicated(['name', 'platform', 'year_of_release'])] 
duplicated
```

```python
data = data.drop(index=[14244, 16230]) # так строк всего 2, принято решение об их удалении по индексу
data.shape[0] # проверяем, что количество строк сократилось на 2
```

```python
data.query('name.isna()')  # смотрим строки с пропусками в столбце 'name'
```

```python
data.dropna(subset=['name'], inplace=True)  
# так как пустых строк 2, и мы не можем установить названия игр, удаляем данные пустые строки
data['name'].isna().sum()  # проверяем
```

```python
data.isna().mean()
```

Количество пустых строк в столбце 'year_of_release' - 269, они составляют 1.6% всех строк. Так как мы не можем достоверно знать год выхода игры (даже если год фигирует в названии игры это не значит, что она выходила именно в этот год), принято решение об удалении данных строк.

```python
data.dropna(subset=['year_of_release'], inplace=True)  # удаляем строки
```

```python
data.isna().sum()  # еще раз смотрим количество пропусков в столбцах
```

* В столбце user_score - 6605 пропусков (40%). Удаление такого количества строк нецелесообразно. TBD в контексте оценки игр означает, что игра ещё не оценена достаточным количеством пользователей. Соответственно, мы не знаем точный рейтинг данных игр и получить эту информацию не можем. Применение медианы и среднего значения недопустимы для рейтингов. В связи с этим, оставляем NaN. Для проведения дальнейших математических расчетов необходимо заменить 'tbd' на NaN.
* В столбце critic_score - 8461 пропусков (51%). Удаление такого количества строк нецелесообразно. Применение медианы и среднего значения недопустимы для рейтингов. Принято решение об оставлении NaN.
* В столбце rating - 6766 пропусков (40%). Удаление такого количества строк нецелесообразно. Определить рейтинг от организации ESRB на основании представленных данных не представляется возможным. Принято решение о замене пропусков на значение 'не определено (undef (undefined))'.

```python
data.replace({'tbd': np.NaN}, inplace=True)  # меняем tbd на NaN
```

```python
data['rating'] = data['rating'].fillna('undef')  # заполняем пустые значения
data['rating'].isna().sum()  # проверяем
```

Для удобства последующего анализа меняем типы данных в столбцах: 'year_of_release' на int, 'critic_score' на float, 'user_score' на float.

```python
data = data.astype(
    {'year_of_release': 'int', 'critic_score': 'float', 'user_score': 'float'}
)
# меняем тип данных в столбцах
```

```python
data.info()
```

```python
# создаем столбец с мировыми продажами
cols = ['na_sales', 'eu_sales', 'jp_sales', 'other_sales']  # определяем столбцы для суммирования
data['world_sales'] = data[cols].sum(axis=1)  # суммируем столбцы
data.head(10)  # проверяем
```

* Привели названия столбцов к нижнему регистру, удалили 2 строки с пропусками в столбце 'name'
* Удалили 269 строк с пустыми значаениями в столбце 'year_of_release'
* В столбце user_score - для проведения дальнейших математических расчетов заменили 'tbd' на NaN
* В столбце critic_score - принято решение об оставлении NaN
* В столбце rating - произвели замену пропусков на значение 'не определено (undef (undefined))'
* Изменили типы данных в столбцах: 'year_of_release' на int, 'critic_score' на float, 'user_score' на float
* Создали отдельный столбец с мировыми продажами (по всем регионам) игр в датпсете

### Продажи по платформам

Посмотрите, как менялись продажи по платформам. Выберите платформы с наибольшими суммарными продажами и постройте распределение по годам. За какой характерный срок появляются новые и исчезают старые платформы?

```python
# строим сводную таблицу по продажам игр по платформам
data.pivot_table(index='platform', values='world_sales', aggfunc='sum').sort_values(by='world_sales', ascending=False)
```

```python
# визуализация данных
(
data.pivot_table(index='platform', values='world_sales', aggfunc='sum')
    .sort_values(by='world_sales', ascending=False)
    .plot(kind='bar' , grid=True, figsize=(10, 5))
)
plt.title('Кол-во мировых продаж игр в зависимости от платформы')
plt.xlabel('Платформа')
plt.ylabel('Продажи')
plt.legend(['проданные копии игр'])
plt.show()
```

```python
# визуализация данных
platform_years = data.pivot_table(index='year_of_release', columns='platform', values='world_sales', aggfunc='sum')
plt.figure(figsize=(25, 15))
sns.heatmap(platform_years, annot=True, annot_kws={'fontsize':8}, cmap='Spectral', center=0)
plt.title('Продажи игр по платформам')
plt.xlabel('Платформа')
plt.ylabel('Год')
plt.show()
```

Как видно из представленных графиков наибольшее количество продаж было у 6 консолей - PS2(1233.56), X360(961.24), PS3(931.34), Wii(891.18), DS(802.78), PS(727.58). Построим для них распределение по годам.

```python
# создаем список с названиями топ 6 платформ с макимальными продажами
top6_platforms = data.pivot_table(index='platform', values='world_sales', aggfunc='sum').sort_values(by='world_sales', ascending=False).reset_index().head(6)['platform'].tolist()
top6_platforms
```

```python
# с помощью цикла перебираем названия платформ и строим визуализацию
for platform in top6_platforms:
    (data.query('platform == @platform')
    .pivot_table(index = 'year_of_release', values = 'world_sales', aggfunc = 'sum')
    .sort_values('year_of_release', ascending = False)['world_sales']
    .plot(figsize = (10, 5), grid=True, label=platform)
    )
    plt.title('Распределение по годам топ-6 платформ')
    plt.xlabel('Год выпуска')
    plt.ylabel('Продажи игр')
    plt.legend() 
```

На визуализации видно, что имеется ошибка в данных и для консоли DS есть игры выпущенные в 1985 году. Найдем эти ошибки и исправим их. 

```python
data.query('platform == "DS" and year_of_release < 2004') # находим игры с годом выпуска ранее Nintendo DS
```

Поиск не дал достоверной информации о дате выхода игры (данные варьируются от 2007 года до 2010 года), в связи с чем принято решение об удалении данной игры из датафрейма.

```python
data = data.drop(15957) # удаляем строку по индексу
data.shape[0] # проверяем, что количество строк уменьшилось на 1
```

```python
# повторно строим визуализацию
for platform in top6_platforms:
    (data.query('platform == @platform')
    .pivot_table(index = 'year_of_release', values = 'world_sales', aggfunc = 'sum')
    .sort_values('year_of_release', ascending = False)['world_sales']
    .plot(figsize = (12, 6), grid=True, label=platform)
    )
    plt.title('Распределение по годам топ-6 платформ')
    plt.xlabel('Год выпуска')
    plt.ylabel('Продажи игр')
    plt.legend() 
```

**Вывод:** из представленного графика видно, что время от начала продаж до конца поддержки консоли составляет около 10 лет. Наибольший пик продаж игр для консолей приходится на 3-5 год жизни консоли. На данном графике из 6 консолей только 3 представляют собой преемственные покололения (PS, PS2, PS3). Соответсвенно, основываясь, на полученных данных период выхода новых консолей составялет 5-6 лет.

### Данные за актуальный период

*Возьмите данные за соответствующий актуальный период. Актуальный период определите самостоятельно в результате исследования предыдущих вопросов. Основной фактор — эти данные помогут построить прогноз на 2017 год.*

```python
relevant_data = data.query('year_of_release > 2013') # фильтруем датафрейм
relevant_data.reset_index(drop=True, inplace=True) # сбрасываем индекс
relevant_data.head() # проверяем
```

```python
# строим общую визуализацию для оценки
plt.figure(figsize=(12, 8))
sns.barplot(
    x='year_of_release', 
    y='world_sales', 
    data=relevant_data, 
    palette='bright',
    hue='platform')
plt.title('Продажи игр на консолях в 2014-2016 годах')
plt.xlabel('Платформа, год')
plt.ylabel('Продажи')
plt.show()
```

```python
# группируем данные по необходимым признакам
relevant_data_grouped = relevant_data.groupby(['platform', 'year_of_release'])['world_sales'].sum().unstack().fillna(0)
relevant_data_grouped
```

```python
# визуализация сгруппированных данных
relevant_data_grouped.plot(kind='bar', grid=True, figsize=(10, 5))
plt.title('Продажи игр на актуальных платформах')
plt.xlabel('Платформа, год')
plt.ylabel('Продажи')
plt.show()
```

**Вывод:** из предоставленных данных следует, что наиболшее количество продаж игр за представленный период наблюдается у платформ PS4, XOne, 3DS. Данные консоли являются наиболее актуальными на рынке на представленный момент и, как было установлено ранее, останутся актуальными и потенциально прибыльными минимум на протяжении 2-3 лет. В 2015 году росли продажи у PS4 и XOne (118.90 и 60.14, соответсвенно), в 2014 году у 3DS (43.76). Однако, в 2015 году отмечен спад продаж игр на всех платформах, возможно из таких факторов как конкуренция, изменение предпочтений потребителей и появление новых технологий.

### Влияние на продажи внутри одной популярной платформы отзывов пользователей и критиков

*Посмотрите, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Постройте диаграмму рассеяния и посчитайте корреляцию между отзывами и продажами. Сформулируйте выводы.*

В качестве примера влияния отзывов пользователей и критиков возьмем наиболее попуряную платформу за актуальный период - PS4.

```python
relevant_data_PS4 = relevant_data.query('platform == "PS4"') # выполняем срез по платформе
relevant_data_PS4[['critic_score', 'user_score', 'world_sales']].corr() # смотрим корреляцию в интересующих столбцах
```

```python
# строим матрицу корреляций
pd.plotting.scatter_matrix(relevant_data_PS4[['critic_score', 'user_score', 'world_sales']], figsize=(12, 12)) 
plt.title('Матрица корреляций PS4')
plt.show()
```

```python
# строим матрицу корреляций на основе тепловой карты
PS4_corr = relevant_data_PS4[['critic_score', 'user_score', 'world_sales']].corr()

plt.figure(figsize=(7, 5))
sns.heatmap(PS4_corr, annot=True, cmap='coolwarm', center=0)
plt.title('Матрица корреляций PS4')
plt.show()
```

**Вывод:** по предоставленным данным отзывы критиков положительно влияют на продажи, а отзывы пользователей и продажи практически не имеют корреляции. Проверим выводы на актуальной платформе конкурентов.

Рассмотрим продажи на других платформах.

```python
# создадим список из 5 наиболее популяных платформ за исключением PS4
top_relevant_platforms = relevant_data.pivot_table(
    index='platform', values='world_sales', aggfunc='sum'
    ).sort_values(by='world_sales', ascending=False).reset_index().head(6)['platform'].tolist()
display(top_relevant_platforms) # проверяем
top_relevant_platforms.pop(0) # удаляем PS4 из списка
top_relevant_platforms # проверяем
```

```python
# применяем цикл для вывода матрицы корреляций и диаграмм рассеяния по оценку пользователей и критиков и продажам
for platform in top_relevant_platforms:
    print(
        f'Матрица корреляций между оценками пользователей, оценками критиков и продажами по {platform} =\n',
        relevant_data.query('platform == @platform')[
            ['user_score', 'critic_score', 'world_sales']
        ].corr(),
    )
    relevant_data.query('platform == @platform').plot(
        x='user_score', y='world_sales', kind='scatter', figsize=(10, 5)
    )
    plt.title(
        f'Диаграмма рассеяния между оценками пользователей и продажами по {platform}'
    )
    plt.xlabel('Оценки пользователей')
    plt.ylabel('Продажи')
    plt.legend([platform])
    plt.show()
    relevant_data.query('platform == @platform').plot(
        x='critic_score', y='world_sales', kind='scatter', figsize=(10, 5), color='red'
    )
    plt.title(
        f'Диаграмма рассеяния между оценками критиков и продажами по {platform}'
    )
    plt.xlabel('Оценки критиков')
    plt.ylabel('Продажи')
    plt.legend([platform])
    plt.show()
```

**Вывод:** согласно полученным данным на продажи игр, а соответственно и консолей, влияют оценки критиков. Однако, наблюдаемая корреляция не слишком сильная (~0,4). Отзывы пользователей практически не влияют на продажу игр и консолей, а иногда корреляция отрицательная. Слабую положительнуая корреляцию можно отметить только у консолей Nintendo. Полученные выводы могут обусловлены большим количеством пропусков в данных, выявленных на этапе предобработки. Соответсвенно, окончательные выводы можно получить только после полного анализа всех данных.
* По количеству продаж наиболее популярным жанром является Action (199.36 игр продано в этом жанре). Далее по количеству продаж распределены Shooter (170.94), Sports (109.48), Role_playing (101.44). Количество продаж в данных жанров выше среднего (61) по всем играм. Однако, при оценке медианного значения продаж на первое место выходят такие жанры как Shooter и Sports, которые стабильно приносят высокий доход.
* Таким образом, в качестве прогноза на 2017 год - наиболее актуальными платформами являются - PS4 и XOneб а наиболее популярными жанрами игр являются - Action и Shooter.

**Портрет пользователя каждого региона**
* Cреди консолей PS4 занимает лидирующую позицию среди данных топ 5 в NA и EU (39% и 55%, соответственно), XOne на втором месте. В JP по всей видимости ппредпочтение отдают портативным консолям. Первое место в топе зпнимает 3DS (48%). На диаграмме также присутсвуеи PSV(16%), несмотря на отсутсвие данной консоли в рейтингах на других рынках.
* Жанр Action является популярным жанром, которые занимает значительную часть рынка видеоигр в данном топе (NA - 29%, EU - 33%, JP - 38%). Жанр Shooter популярен в NA (32%) и EU (29%), но менее популярен в JP. В NA и EU популярен жанр Sports, в отличие от JP. Также можно выделить жанр Role-Playing, которые присутсвует в топах всех 3 регионах.
* В NA и EU лучше всего продаются игры с рейтингом М. В JP на первом месте по продпжпм игры с рейтингом T. По всей видимости это связано с разной возрастной аудиторией игр.
